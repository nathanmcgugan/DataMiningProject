# Joining CSVs into a feather file (~1.3 GB)

```python
import pandas as pd

DATA_SET_DIR = './instacart-market-basket-analysis'

orders = pd.read_csv(f'{DATA_SET_DIR}/orders.csv')
orders
```

```python
# The 'prior' csv contains all the products ordered in all orders
order_products = pd.read_csv(f'{DATA_SET_DIR}/order_products__prior.csv')
order_products
```

```python
orders = pd.merge(orders, order_products, on='order_id', how='left')

orders
```

```python
del order_products

products = pd.read_csv(f'{DATA_SET_DIR}/products.csv')
orders = pd.merge(orders, products, on='product_id', how='left')
orders
```

```python
del products

aisles = pd.read_csv(f'{DATA_SET_DIR}/aisles.csv')
orders = pd.merge(orders, aisles, on='aisle_id', how='left')

orders
```

```python
del aisles

departments = pd.read_csv(f'{DATA_SET_DIR}/departments.csv')
orders = pd.merge(orders, departments, on='department_id', how='left')

orders
```

```python
orders.to_feather('products_and_orders.feather')
```
